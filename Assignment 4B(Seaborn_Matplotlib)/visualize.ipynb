```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
sb.get_dataset_names()
```

```python
import pandas as pd
```

```python
data1 = pd.read_csv('forestfires.csv')
```

```python
data1.head()
```

```python
sb.countplot(data=data1, x='wind', palette='pastel')
```

```python
sb.countplot(data=data1, y='day', palette='pastel')
```

```python
data1.columns
```

```python
sb.countplot(data=data1, y='day', palette='pastel', hue='month')
plt.title('Countplot of Relative Humidity')
```

```python
data2 = pd.read_csv('AirQuality.csv',delimiter=';')
```

```python
data2.head()
```

```python
data2.columns
```

```python
sb.histplot(data=data1, x='temp', hue='X')
plt.title('Histplot of Temperature')
```

```python
sb.boxenplot(data=data2, x='T')
plt.title('Boxenplot of Relative Humidity')
```

```python
sb.violinplot(
    data=data1,
    x='temp',
)
plt.title('Violinplot of Relative Humidity')
```

```python
sb.rugplot(data=data2, x='CO(GT)')
plt.title('Rugplot of CarbonMonoxide')
```

```python
plt.title('Kdeplot of Temperature')
sb.kdeplot(data=data1, x='temp')
```

```python
plt.title('Kdeplot of Temperature')
sb.kdeplot(data=data1, x='temp', shade=True)
```

```python
plt.title('Kdeplot of Temperature')
sb.kdeplot(data=data1, x='temp', hue='day')
```

```python
plt.title('Stripplot of Wind')
sb.stripplot(data=data1, x='wind')
```

```python
sb.swarmplot(data=data1, x='temp')
plt.title('Swarmplot of Temperature')
```

```python
sb.ecdfplot(data=data1, x='temp')
plt.title('ecdfplot of Temperature')
```

```python
sb.ecdfplot(data=data1, x='temp', hue='day')
plt.title('ecdfplot of Temperature')
```
